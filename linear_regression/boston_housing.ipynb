```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston
```

```python
boston_dataset = load_boston()
boston_dataset.keys()
```

```python
boston_data = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston_data.head()
```

```python
boston_data["TARGET"] = boston_dataset.target
boston_data
```

```python
boston_data.isnull().any()
```

```python
# set the size of the figure
sns.set(rc={'figure.figsize':(10,8.27)})

# plot a histogram showing the distribution of the target values
# sns.distplot(boston_data['AGE'], bins=100)
sns.distplot(boston_data['TARGET'], bins=100)

plt.show()
```

```python
correlation_matrix = boston_data.corr().round(2)
correlation_matrix
```

```python
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
features = ["RM", "LSTAT"]
# features = boston_dataset.feature_names
for i, col_name in enumerate(features):
    plt.subplot(1, len(features), i+1)
    x = boston_data[col_name]
    y = boston_data["TARGET"]
    plt.scatter(x, y, marker="o")
    plt.title(col_name)
    plt.xlabel(col_name)
    plt.ylabel('MEDV')
```

```python
# Prepare data for training
X = pd.DataFrame(np.c_[boston_data["LSTAT"], boston_data["RM"]], columns=["LSTAT", "RM"])
y = boston_data["TARGET"]
X
```

```python
# Split data into traning and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=8)
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

```python
# model evaluation for training set
y_train_predict = linear_model.predict(X_train)
rmse = np.sqrt(mean_squared_error(y_train, y_train_predict))
r2 = r2_score(y_train, y_train_predict)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

```python
y_test_predict = linear_model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_test_predict))
r2 = r2_score(y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

```python
plt.scatter(y_test, y_test_predict, marker='o')
plt.xlabel("y_test")
plt.ylabel("y_test_predict")
plt.show()
```
